```python
x=data.drop(['price_range'],axis=1)
x
```

```python
y=data[['price_range']]
y
```

```python
trainx,testx,trainy,testy = train_test_split(x,y,test_size = 0.2)
```

# Gaussian

```python
gnb=GaussianNB().fit(trainx,trainy)
gnb.predict(testx)
gnb.score(testx,testy)
```

# Bernoulli's

```python
bnb=BernoulliNB().fit(trainx,trainy)
bnb.predict(testx)
bnb.score(testx,testy)
```

# Multinomial

```python
mnb=MultinomialNB().fit(trainx,trainy)
mnb.predict(testx)
mnb.score(testx,testy)
```
